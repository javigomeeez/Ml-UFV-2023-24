```python
import torch
import pandas as pd
#Tensor de ceros
def create_zeros(shape):
   
    return torch.zeros(shape)
#tensor de unos

def create_ones(shape):

    return torch.ones(shape)

#tensor numeros random

def create_random(shape):

    return torch.rand(shape)

#tensor suma

def add_tensors(tensor1, tensor2):

    return tensor1 + tensor2

#tensor multiplicación
def multiply_tensors(tensor1, tensor2):

    return tensor1 * tensor2

#tensor de normalización de valores , util para el tratamiento de desviaciones tipicas altas que no interesan para ela anlisis
def normalize_tensor(tensor):

    mean = tensor.mean()
    std = tensor.std()
    return (tensor - mean) / std

#Limitador de valores 

def clamp_tensor(tensor, min_value, max_value):

    return torch.clamp(tensor, min_value, max_value)
```

# Caso de uso practico basado en ejercicio de analisis financiero

```python
# Creamos  tensor con indicadores  de clientes
indicadores_financieros = create_random((100, 10))  

# Normalizamos los indicadores para poder tratarlos correctamente
indicadores_norm = normalize_tensor(indicadores_financieros)

# Creamos un tensor de factores de estrés económico,es decir estresamos los factores produciendo una caida del 50%
factores_estres = create_random((10,)) - 0.5 

# Aplicamos los factores de estrés 
indicadores_estresados = multiply_tensors(indicadores_norm, factores_estres)

# Limitamos los valores de los indicadores estresados , esto es con el fin de impedir outliers
indicadores_clamp = clamp_tensor(indicadores_estresados, min_value=-2, max_value=2)

# Creamos un tensor con las puntuaciones crediticias de los clientes
puntuaciones_credito = create_random((100, 1)) * 600 + 300 

# Actualizamos las puntuaciones con los factores estresados
puntuaciones_actualizadas = add_tensors(puntuaciones_credito, indicadores_clamp.sum(dim=1, keepdim=True))

# Limitamos las puntuaciones dentro de un rango (esto no se como de veridico es , he cogido los valores que he encontrado pero pueden ser no exactamente estos)
puntuaciones_finales = clamp_tensor(puntuaciones_actualizadas, min_value=300, max_value=900)

# Lo pasamos a df para tenerlo ordenado y asi poder manejarlo mas facilmente
puntuaciones_finales_np = puntuaciones_finales.numpy()
df_puntuaciones_finales = pd.DataFrame(puntuaciones_finales_np, columns=['Puntuación de Crédito'])

#Supongamos un umbral de aprobación del credito de 650
umbral_aprobacion = 650

# Creamos una columna para saber si se ha aprobado o no y visualizamos
df_puntuaciones_finales['Crédito'] = df_puntuaciones_finales['Puntuación de Crédito'].apply(lambda x: 'Aprobado' if x >= umbral_aprobacion else 'Denegado')


df_puntuaciones_finales
```
